```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('laptops_train.csv',usecols=['Manufacturer','Category','Screen Size','CPU','RAM',' Storage','Operating System','Price'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Operating System'].value_counts()
```

```python
df['Operating System']=np.where(df['Operating System']=='macOS','Mac OS',df['Operating System'])
```

```python
df['Operating System']=np.where(df['Operating System']=='No OS','Windows',df['Operating System'])
```

```python
df.head()
```

```python
df['Operating System'].value_counts()
```

```python
df['Screen Size'] = df['Screen Size'].str.replace('"',"")
df['Screen Size'] = df['Screen Size'].astype('float')
```

```python
df['RAM'] = df['RAM'].str.replace('GB',"")
df['RAM'] = df['RAM'].astype('int')
```

```python
df.head(25)
```

```python
df['Category'].value_counts()
```

```python
df['CPU'] = df['CPU'].apply(lambda x:" ".join(x.split()[0:3]))
df.head()
```

```python
def fetch_processor(text):
    if text=="Intel Core i7" or text=="Intel Core i5" or text=="Intel Core i3":
        return text
    else:
        if text.split()[0] == "Intel":
            return "Other Intel Processor"
        else:
            return "AMD Processor"
```

```python
df['CPU'] = df['CPU'].apply(fetch_processor)
df.head()
```

```python
df[' Storage']=df[' Storage'].astype(str).str[0:3]
```

```python
df[' Storage'].value_counts()
```

```python
ordinal=df.groupby(['Category'])['Price'].mean().sort_values().index
enumerate(ordinal,0)
ordinal_l={k:i for i,k in enumerate(ordinal,0)}
df['Category']=df['Category'].map(ordinal_l)
df.head()
ordinal_l
```

```python
ordinal=df.groupby(['CPU'])['Price'].mean().sort_values().index
enumerate(ordinal,0)
ordinal_l={k:i for i,k in enumerate(ordinal,0)}
df['CPU']=df['CPU'].map(ordinal_l)
df.head()
ordinal_l
```

```python
ordinal=df.groupby(['Operating System'])['Price'].mean().sort_values().index
enumerate(ordinal,0)
ordinal_l={k:i for i,k in enumerate(ordinal,0)}
df['Operating System']=df['Operating System'].map(ordinal_l)
df.head()
ordinal_l
```

```python
ordinal=df.groupby(['Screen Size'])['Price'].mean().sort_values().index
enumerate(ordinal,0)
ordinal_l={k:i for i,k in enumerate(ordinal,0)}
df['Screen Size']=df['Screen Size'].map(ordinal_l)
df.head()
ordinal_l
```

```python
ordinal=df.groupby(['RAM'])['Price'].mean().sort_values().index
enumerate(ordinal,0)
ordinal_l={k:i for i,k in enumerate(ordinal,0)}
df['RAM']=df['RAM'].map(ordinal_l)
df.head()
ordinal_l
```

```python
lst_10=df.Manufacturer.value_counts().sort_values(ascending=False).head(10).index
lst_10=list(lst_10)
for categories in lst_10:
    df[categories]=np.where(df['Manufacturer']==categories,1,0)
df.head()
```

```python
lst_10=df[' Storage'].value_counts().sort_values(ascending=False).head(10).index
lst_10=list(lst_10)
for categories in lst_10:
    df[categories]=np.where(df[' Storage']==categories,1,0)
df.head()
lst_10
```

```python
df.head()
```

```python
df.info()
```

```python
X= df.drop(columns=['Manufacturer',' Storage','Price'])
y= df['Price']
```

```python
X.head(10)
```

```python
y.head(10)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X, y, test_size = 0.1, random_state = 0)
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
regr =AdaBoostRegressor(DecisionTreeRegressor(max_depth=4), n_estimators=300, random_state=0)
regr.fit(X_train, y_train)
```

```python
regr.predict(X_test)
```

```python
y_test
```

```python
p=regr.score(X_train, y_train)
p
```
